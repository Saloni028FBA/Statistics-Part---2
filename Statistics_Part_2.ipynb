## Theory Part

1. What is hypothesis testing in statistics?

   - Hypothesis testing is a fundamental concept in statistics used to make inferences about a population based on sample data.
   
   - It helps determine whether a certain assumption (hypothesis) about the population is likely to be true.

   - Here's how it works:

     1. State the Hypotheses:
     
         - Null Hypothesis (H₀): This is the default assumption that there is no effect or no difference.
         
         - Alternative Hypothesis (H₁ or Hₐ): This is the statement that contradicts the null hypothesis and represents what the researcher wants to prove.

    2. Choose a Significance Level (α):
    
        - Commonly set at 0.05 (5%), meaning there’s a 5% risk of incorrectly rejecting the null hypothesis.

        - If p-value ≤ α, reject H₀.
        
        - If p-value > α, fail to reject H₀.

    3. Select a Test Statistic and Conduct the Test:
    
       - Depending on the type of data, different tests are used (e.g., t-test, chi-square test, ANOVA, z-test).

    4. Types of Errors:
    
        - Type I Error: Rejecting H₀ when it's actually true.
        
        - Type II Error: Failing to reject H₀ when it's actually false.

    5. Calculate the p-value:

        - The p-value indicates the probability of obtaining results at least as extreme as the observed ones, assuming the null hypothesis is true.

        -  A small p-value (typically < 0.05) means there is strong evidence against H₀

    6. Make a Decision:
    
        - If p-value < α, reject the null hypothesis (suggests strong evidence for the alternative hypothesis).
        
        - If p-value ≥ α, fail to reject the null hypothesis (not enough evidence to support the alternative hypothesis).

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

   - The null hypothesis (H₀) and alternative hypothesis (H₁ or Hₐ) are the two opposing statements in hypothesis testing.

   - Null Hypothesis (H₀):
   
     - It is the default or starting assumption in hypothesis testing.
     
     - It claims that there is no effect, no difference, or no relationship between variables.
     
     - You assume it's true unless (Try to find evidence against it) the data gives strong evidence against it.

     - If p-value is low = Reject H₀

     - It suggests that any observed differences in the data are due to chance.
     
     - Example: In a drug trial, H₀ could state that "The new drug has no effect on blood pressure."

  - Alternative Hypothesis (H₁ or Ha):
  
    - This is the opposite of the null hypothesis.
    
    - It states that there is an effect, a difference, or a relationship.
    
    - You try to find evidence to support this hypothesis.

    - If p-value is low = Support H₁

    - It contradicts the null hypothesis.
    
    - Example: For the same drug trial, H₁ could state that "The new drug significantly lowers blood pressure."

3. What is the significance level in hypothesis testing, and why is it important?

   - The significance level, denoted by the Greek letter α (alpha), is the threshold you set before conducting a hypothesis test to decide how much risk of error you're willing to accept when rejecting the null hypothesis (H₀).

   - α is the probability of making a Type I error, which occurs when we falsely conclude there is an effect or difference. (means rejecting H₀ when it is actually true.)

   - Common values of α are:
   
      - 0.05 (5%)
      
      - 0.01 (1%)
      
      - 0.10 (10%)

   - It is important because of the following reasons:

      1. Determines decision criteria:

         - If p-value < α, we reject H₀ (suggests strong evidence for the alternative hypothesis H₁).
         
         - If p-value ≥ α, we fail to reject H₀ (not enough evidence to support H₁).

      2. Controls the probability of incorrect rejection:
      
         - A lower significance level (e.g., 0.01) makes it harder to reject H₀ and reduces false positives.
         
         - A higher significance level (e.g., 0.10) makes it easier to reject H₀, increasing the risk of false positives.

      3. Balances Type I and Type II errors:
      
         - A very low α (e.g., 0.001) minimizes false positives but may increase Type II errors (false negatives).
         
         - Choosing α = 0.05 is common in practice—it provides a reasonable trade-off.

4. What does a P-value represent in hypothesis testing?
   
   - The p-value (probability value) is one of the most important concepts in hypothesis testing. It quantifies the strength of the evidence against the null hypothesis (H₀) based on the observed sample data.

   - The p-value is the probability of obtaining a test statistic at least as extreme as the one observed, assuming that the null hypothesis (H₀) is true.

   - It helps determine whether the evidence is strong enough to reject H₀ in favor of the alternative hypothesis (H₁).

   - Key Interpretations:

      - A small p-value (≤ α, e.g., ≤ 0.05) suggests strong evidence against H₀, leading to its rejection.
      
      - A large p-value (> α, e.g., > 0.05) means weak evidence against H₀, so we fail to reject H₀.

   - Example: Coin Toss (Testing Fairness)
   
      - H₀: The coin is fair (P(Heads) = 0.5).
      
      - H₁: The coin is biased (P(Heads) ≠ 0.5).
      
      - Experiment: Flip the coin 100 times, get 60 heads.
      
      - P-value calculation: Probability of getting ≥60 or ≤40 heads if the coin is fair.
      
      - Suppose p = 0.04.
      
      - Conclusion: Since p < 0.05, reject H₀ (evidence suggests the coin is biased).

   -  In Simple Terms:
   
       - Small p-value = strong evidence against H₀
       
       - Big p-value = weak evidence; can't reject H₀

5. How do you interpret the P-value in hypothesis testing?

    - The P-value in hypothesis testing helps you determine the strength of your evidence against a null hypothesis. Here’s how to interpret it:
    
       - A low P-value (typically ≤ 0.05) suggests that the observed data is unlikely under the null hypothesis, meaning you have strong evidence to reject the null hypothesis.
       
       - A high P-value (greater than 0.05) indicates that the data is consistent with the null hypothesis, meaning you don’t have strong enough evidence to reject it.

    -  Step-by-Step Interpretation:

        - Step 1: Set a Significance Level (α)
        
          - Before testing, choose a threshold (commonly α = 0.05). This is the maximum acceptable probability of a Type I error (falsely rejecting H₀).

        - Step 2: Compare P-value to α

          - If p ≤ α → Reject H₀ (statistically significant result).
          
             - Example: p = 0.03 (≤ 0.05) → Reject H₀.
             
          - If p > α → Fail to reject H₀ (not statistically significant).
          
             - Example: p = 0.12 (> 0.05) → Cannot reject H₀.

    - Step 3: Practical Conclusion
    
       - Rejecting H₀: Suggests the alternative hypothesis (H₁) may be true.
       
       - Failing to reject H₀: Does not prove H₀ is true; it just means there’s insufficient evidence against it.

    - Example: A/B Testing (Website Conversions)
    
       - H₀: New webpage design has no impact on conversion rates.
       
       - H₁: New design increases conversions.
       
       - Result: p = 0.20 (α = 0.05).
       
       - Interpretation: Since p > α, fail to reject H₀. No strong evidence the new design works.

6. What are Type 1 and Type 2 errors in hypothesis testing?

   - Type 1 and Type 2 errors are two kinds of mistakes that can occur in hypothesis testing:

     - Type 1 Error (False Positive, α):

       - This happens when you reject a true null hypothesis.

       - In other words, you conclude there is an effect when there actually isn’t one.

       - The probability of making a Type 1 error is called alpha (α), and it's usually set at 0.05 (5%).

       - Example: A medical test incorrectly detects a disease in a healthy person.

    - Type 2 Error (False Negative, β):
    
      - This occurs when you fail to reject a false null hypothesis.
      
      - It means you miss detecting an actual effect.
      
      - The probability of making a Type 2 error is called beta (β), and it depends on factors like sample size and effect size.
      
      - Example: A medical test fails to detect a disease in a person who actually has it.

   - In hypothesis testing, reducing one type of error often increases the other, so researchers balance them carefully based on the consequences of errors in their analysis.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

  - In hypothesis testing, the primary difference between a one-tailed test and a two-tailed test lies in the directionality of the alternative hypothesis and, consequently, the location of the critical region(s) used to determine statistical significance.

    - One-Tailed Test (Directional):

      - A one-tailed test is used when there's a specific directional expectation for the effect.
      
      - It tests if a sample statistic is significantly greater than or significantly less than a population parameter, but not both.

      - The alternative hypothesis specifies a direction.
      
         - Right-tailed test: H1 : parameter > value (e.g., a new drug increases recovery rate)
         
         - Left-tailed test: H1 : parameter < value (e.g., a new manufacturing process reduces defects)

      - Critical Region -> The critical region (or rejection region) is the area in the sampling distribution where, if the test statistic falls, the null hypothesis (H0 ) is rejected. The significance level (alpha, α) determines the size of this region.

        - The entire alpha level (α) is allocated to one tail of the distribution (either the right or the left, depending on the alternative hypothesis). This makes the test more powerful at detecting an effect in the specified direction.
        
          - For a right-tailed test, the critical region is entirely in the right tail.
          
          - For a left-tailed test, the critical region is entirely in the left tail.
   
      - When to Use One-tailed test
      
         -  when a strong theoretical basis or prior research suggesting an effect in a specific direction before collecting or analyzing data.
    
         - For example, if you are testing if a new fertilizer increases crop yield, you'd use a right-tailed test. You are only interested in whether it increases the yield, not if it decreases it or simply changes it.

      - Statistical Power
      
         -  A one-tailed test generally has more statistical power to detect an effect in the specified direction compared to a two-tailed test with the same significance level.
        
         - This is because the entire α is concentrated in one tail, making the critical value easier to reach if the effect is indeed in that direction.

   - Two-Tailed Test (Non-directional):

      - A two-tailed test is used when there's no specific directional expectation.
      
      - It tests if a sample statistic is significantly different from a population parameter, meaning it could be either greater than or less than.

      - The alternative hypothesis does not specify a direction; it simply states that there is a difference.

        - H1 : parameter not equals to value (e.g., a new teaching method has a different impact on test scores, either higher or lower).

      - The critical region (or rejection region) is the area in the sampling distribution where, if the test statistic falls, the null hypothesis (H0) is rejected. The significance level (alpha, α) determines the size of this region.

        - In Two-tailed test, the alpha level (α) is split equally between both tails of the distribution (e.g., if α=0.05, then 0.025 is in the left tail and 0.025 is in the right tail). This means the test is looking for an extreme value in either direction.

     - When to Use One-tailed test

        - This is generally the more conservative and commonly used approach, especially when:
        
         - You want to see if there's any difference, regardless of direction.
         
         - There's no strong prior evidence to predict the direction of the effect.
         
         - An effect in the opposite direction to what you might expect would still be an important finding.
         
         - For example, if testing a new drug, you'd want to know if it not only helps but also if it significantly harms (an effect in the opposite direction).

     - Statistical Power

        - A two-tailed test has less power to detect an effect in a specific direction because the α is divided. However, it can detect an effect in either direction.

8. What is the Z-test, and when is it used in hypothesis testing?

   - The Z-test is a type of statistical test used in hypothesis testing to determine whether there is a significant difference between a sample mean and a population mean.
   
   - It is based on the standard normal distribution (Z-distribution) and is particularly useful when the sample size is large (n ≥ 30) and the population standard deviation is known.

   - When to Use the Z-test?
   
      - A Z-test is appropriate when:

         - The data is normally distributed or the sample size is large enough for the Central Limit Theorem to apply.
         
         - The population standard deviation (σ) is known, making it different from the T-test, which is used when σ is unknown.
         
         - when comparing means:
         
            - Comparing a sample mean to a known population mean (One-Sample Z-test).
            
            - Comparing two independent sample means (Two-Sample Z-test).

        - Independent Data Points : The observations in the sample should be independent of each other.

  - Example Use Case
  
     - Suppose a university claims that the average student GPA is 3.5, but you suspect it may be lower. You collect a sample of 50 students, calculate the sample mean GPA, and use a Z-test to determine if there's enough evidence to reject the university’s claim.

     - Formula for Z-test   
     
        𝑍 = 𝑥 − 𝜇 / 𝜎 / squareroot (𝑛)

       - 𝑥 = Sample mean
    
       - 𝜇 = Population mean
     
       - 𝜎 = Population standard deviation
     
       - 𝑛 = Sample size

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

   - The Z-score measures how far a data point or sample mean deviates from the population mean, expressed in terms of standard deviations.
   
   - In hypothesis testing, it's used to determine whether an observed value is significantly different from the expected value.

   - How to Calculate the Z-Score.
   
      - The formula for the Z-score is :

           𝑍 = 𝑥 − 𝜇 / 𝜎 / squareroot (𝑛)

           - 𝑥 = Sample mean
           
           - 𝜇 = Population mean
           
           - 𝜎 = Population standard deviation
           
           - 𝑛 = Sample size

   - What the Z-Score Represents:
   
      - A high positive Z-score (e.g., Z = 2.5) means the sample mean is much greater than the population mean.
      
      - A negative Z-score (e.g., Z = -1.8) means the sample mean is below the population mean.
      
      - A Z-score close to 0 means the sample mean is very similar to the population mean.

  - Application in Hypothesis Testing:
  
     - One-Sample Z-test: Used to test if a sample mean differs significantly from a population mean.
     
     - The Z-score is compared against critical values from the standard normal distribution to decide whether to reject the null hypothesis.

16. How is Bayes' Theorem used in statistics, and what is its significance?

    - Bayes’ Theorem is a fundamental concept in probability theory and statistics that allows us to update the probability of an event based on new evidence.

    - It’s especially powerful in situations where prior knowledge influences the likelihood of outcomes.

    - In statistics, it forms the backbone of Bayesian inference, an approach that contrasts with traditional frequentist statistics (e.g., p-values, confidence intervals).

    - Bayes’ Theorem Formula:
    
      - P( A ∣ B ) =  P ( B ∣ A ) . P(A) / P(B)

        - P(A ∣ B) => Posterior probability (updated belief about A after observing B).

        - P(B ∣ A) => Likelihood (probability of observing B if A is true)

        - P(A): Prior probability (initial belief about A).

        - P(B): Marginal probability (total probability of observing B).

    - Significance of Bayes' Theorem:

        1. Incorporates Prior Knowledge
            
            - Uses existing beliefs (priors) instead of assuming fixed parameters.
            
            - Example: Drug trials can leverage past studies.

        2.  Intuitive Probabilities
        
            - Directly answers: "What’s the probability this hypothesis is true?"
            
            - Contrast: Frequentist p-values ask "How likely is the data if H₀ is true?"

        3.  Handles Small Samples Better:
        
             - Priors stabilize estimates when data is limited.

        4. Continuous Learning
       
             - Posteriors become priors for new data (sequential updating).
             
             - Crucial in predictive modeling and AI applications.

    - How Bayes' Theorem is Used in Statistics

        1. Medical Diagnosis
        
            - Helps calculate the probability of a disease given a positive test result.
            
            - Example: If a medical test is 90% accurate, Bayes’ Theorem refines the probability by incorporating the disease’s actual prevalence.

        2. Spam Filtering in Email Systems
        
            - Used to determine whether an email is spam based on words commonly found in spam messages.
            
            - Gmail and other email services use Bayesian filtering.

        3. Stock Market & Risk Analysis
        
             - Helps update predictions about market trends based on new financial data.

17. What is the Chi-square distribution, and when is it used?

     - The Chi-square (χ²) distribution is a continuous probability distribution that arises in hypothesis testing, primarily for categorical data.

     - It arises when summing the squares of independent standard normal variables, making it particularly useful for hypothesis testing.

     -  It is used to assess goodness-of-fit, independence, and variance in statistical analyses.

     - Key Characteristics of the Chi-square Distribution

         1. It is positively skewed, meaning it is not symmetric like the normal distribution.

         2. The shape depends on degrees of freedom (df) as d𝑓 increases, it becomes more symmetrical.

         3. Values are always non-negative, since it involves squared terms (χ² ≥ 0).

         4. Mean = df and Variance = 2 × df.

     - When is the Chi-square Distribution Used?

         1. Chi-square Goodness-of-Fit Test
         
            - Used to determine whether a observed categorical data matches an expected distribution.
            
            - Example: Checking if dice rolls are fair by comparing observed vs. expected frequencies.

         2. Chi-square Test for Independence
         
            - Used to analyze relationships between two categorical variables.
            
            - Example: Testing if customer preference for a product depends on age group.

         3. Chi-square Test for Variance
         
             - Used to determine whether the variance of a population is equal to a specified value.
             
             - Example: Checking if the variance in machine performance meets industry standards.

18. What is the Chi-square goodness of fit test, and how is it applied?

    - It assesses how well observed data fit a theoretical distribution, making it useful in categorical data analysis.

    - The purpose is to assess whether the observed distribution of a single categorical variable differs significantly from a hypothesized distribution.

        - Oi = Observed frequency
        
        - 𝐸i = Expected frequency

    -  Steps to Apply the Chi-square Goodness of Fit Test

        1. State the Hypotheses:
        
           - Null (H₀): Observed data fits the expected distribution.
           
           - Alternative (H₁): Observed data does not fit the expected distribution.

        2. Calculate Expected Frequencies: Based on theoretical probabilities or previous data.
        
          - For uniform distribution: 𝐸𝑖 = 𝑁 / 𝑘

            - 𝑁 is total observations
            
            - k is number of categories.

        3. Compute the Chi-square statistic using the formula.

            - χ 2 = ∑ (Oi − Ei)2 / Ei

              - Oi = Observed frequency
        
              - 𝐸i = Expected frequency

        4. Find the degrees of freedom (df):
        
           -  𝑑𝑓 = number of categories − 1

        5. Determine the p-value from a Chi-square distribution table using your calculated statistic and degrees of freedom.

        6. Draw a conclusion:

19. What is the F-distribution, and when is it used in hypothesis testing?

    - The F-distribution is a probability distribution commonly used in statistical hypothesis testing, particularly when comparing variances between groups.
    
    - It is derived from the ratio of two Chi-square distributed variables and is always positively skewed (non-negative values).

    - Key Properties of the F-Distribution:

       1. It is used for variance comparisons in hypothesis testing.

       2. The shape depends on two parameters: degrees of freedom (𝑑𝑓1,𝑑𝑓2).

       3. It is asymmetric and right-skewed, meaning values are always positive.

    - When is the F-Distribution Used?

       1. Analysis of Variance (ANOVA)
       
          - Used to test whether multiple group means are significantly different.
          
          - Example: Comparing the effectiveness of three different drug treatments.

       2. F-Test for Variance Comparison
       
          - Used to check if two populations have equal variances.
          
          - Example: Comparing the variability of production defects in two manufacturing plants.

       3. Regression Analysis
       
          - Used in multiple regression to evaluate whether independent variables significantly explain variation in the dependent variable.

    - Formula for the F-Statistic

       - 𝐹 = 𝑆21 / 𝑆22

         - 𝑆21 = Variance of sample 1
         
         - 𝑆22 = Variance of sample 2

20. What is an ANOVA test, and what are its assumptions?

    - The ANOVA test is a statistical method used to compare the means of three or more groups to determine if there is a significant difference between them.
    
    - It helps evaluate whether observed differences are due to real effects or just random variation.

    - Types of ANOVA
    
       - One-Way ANOVA – Compares the means of multiple independent groups based on a single factor.
       
       - Two-Way ANOVA – Examines the impact of two independent variables on the dependent variable.
       
       - Repeated Measures ANOVA – Used when the same group is measured multiple times under different conditions

    - Assumptions of ANOVA

        - Independence –> Observations within each group must be independent of one another.

        - Normality –> The data in each group should follow a normal distribution (can be checked using normality tests like the Shapiro-Wilk test)

        - Homogeneity of Variances –> The variances of all groups should be approximately equal (Levene’s test can be used to check this).

        - Random Sampling –> Data should be randomly collected to avoid biases.

    -  If assumptions are violated:
    
        - Use Welch’s ANOVA (if variances are unequal).
        
        - Use non-parametric tests like Kruskal-Wallis (if data are not normal)

21. What are the different types of ANOVA tests?

    - ANOVA tests are designed to compare group means, but different variations exist depending on the number of factors and the nature of the data.
    
    - Here are the main types:

       1. One-Way ANOVA
       
          - Compares the means of three or more independent groups based on one factor.
          
          - Example: Comparing the average test scores of students from different schools.

       2. Two-Way ANOVA
       
          - Compares the effect of two independent variables on a dependent variable.
          
          - Example: Analyzing how diet and exercise affect weight loss.

       3. Repeated Measures ANOVA
       
           - Used when the same group is tested under different conditions multiple times.
           
           - Example: Measuring blood pressure before, during, and after taking medication.

       4. Mixed-Design ANOVA
       
           - Combines repeated measures with independent variables.
           
           - Example: Studying cognitive performance over time while comparing different age groups.

       5. MANOVA (Multivariate ANOVA)   
       
           - Extends ANOVA to compare multiple dependent variables at the same time.
           
           - Example: Examining how teaching methods affect both math and reading scores.

       6. ANCOVA (Analysis of Covariance)
       
           - Adjusts for the influence of covariates that might affect the dependent variable.
           
           - Example: Comparing student performance while controlling for previous knowledge.

    - Each type serves different research needs, helping analyze variability across different groups and conditions.

22. What is the F-test, and how does it relate to hypothesis testing?

    - The F-test is a statistical test used to compare variances between groups and determine whether they are significantly different.
    
    - It is closely tied to hypothesis testing, particularly in cases involving regression analysis and ANOVA.

    - F = Variance 1 / Variance 2    Or more generally

    - F = Between-group variance / Within-group variance

    - If the F-ratio is much greater than 1, it suggests the groups differ more than would be expected by chance.

    - How It Relates to Hypothesis Testing:
    
       1. Analysis of Variance (ANOVA)
       
           - Determines if multiple group means differ significantly.
           
        2. Regression Analysis
        
           - Evaluates whether independent variables explain variation in the dependent variable.
           
        3. Comparing Variances
        
            - Used to test if two samples have equal variances before applying other statistical tests.

```python
# 9. Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_z_scores(data):
    """
    Calculates Z-scores for a given dataset.

    Parameters:
    - data: A list or NumPy array of numerical values

    Returns:
    - Z-scores as a NumPy array
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

def plot_z_scores(z_scores):
    """
    Plots a histogram of Z-scores with a standard normal curve overlay.

    Parameters:
    - z_scores: A NumPy array of Z-scores
    """
    plt.hist(z_scores, bins=20, density=True, alpha=0.7, color='blue', edgecolor='black', label="Z-score Distribution")

    # Overlay standard normal curve
    x = np.linspace(-3, 3, 100)
    plt.plot(x, stats.norm.pdf(x), color='red', linestyle='dashed', linewidth=2, label="Standard Normal PDF")

    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.title("Z-score Histogram with Standard Normal Curve")
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Example usage
data = np.random.normal(50, 10, 1000)  # Generate random dataset
z_scores = calculate_z_scores(data)
plot_z_scores(z_scores)
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats

# Sample dataset
sample_data = np.array([48000, 52000, 51000, 49500, 50500, 49000, 53000, 47000, 50000, 52500])
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
n = len(sample_data)
pop_mean = 50000  # Hypothesized population mean

# Calculate Z-score
z_score = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))

# Compute p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference detected.")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as stats

# Sample dataset (e.g., test scores)
data = np.array([78, 85, 90, 92, 88, 76, 81, 95, 89, 84])

# Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
n = len(data)
confidence_level = 0.95

# Compute margin of error using the t-distribution (for small samples)
t_score = stats.t.ppf((1 + confidence_level) / 2, df=n-1)  # Critical value
margin_of_error = t_score * (sample_std / np.sqrt(n))

# Confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Interpretation
print(f"We are {confidence_level * 100:.0f}% confident that the true population mean falls between {lower_bound:.2f} and {upper_bound:.2f}.")
```

```python
# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
import scipy.stats as stats

def generate_data_and_confidence_interval(mean=50, std_dev=10, size=100, confidence=0.95):
    # Step 1: Generate random sample data
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Step 2: Compute sample mean and standard error
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # use ddof=1 for sample std
    standard_error = sample_std / np.sqrt(size)

    # Step 3: Calculate confidence interval
    z_critical = stats.norm.ppf((1 + confidence) / 2)  # two-tailed Z value
    margin_of_error = z_critical * standard_error

    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error

    # Step 4: Print results
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
    print(f"This means we are {int(confidence * 100)}% confident that the true population mean lies within this interval.")

# Example usage
generate_data_and_confidence_interval()
```

```python
# 19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution

import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

def poisson_cdf_example(lmbda=4, k=6):
    """
    Calculate and interpret the CDF of a Poisson distribution.

    Parameters:
    - lmbda: The expected number of occurrences (λ)
    - k: The value up to which the CDF is calculated
    """
    # Step 1: Calculate the CDF
    cdf_value = poisson.cdf(k, mu=lmbda)

    # Step 2: Interpret the result
    print(f"P(X ≤ {k}) where X ~ Poisson(λ={lmbda}) = {cdf_value:.4f}")
    print(f"This means there is a {cdf_value * 100:.2f}% chance that the event occurs {k} or fewer times.")

    # Step 3: Optional - Visualize the CDF
    x = np.arange(0, k + 10)
    y = poisson.cdf(x, mu=lmbda)

    plt.figure(figsize=(8, 5))
    plt.step(x, y, where='post', label=f'CDF of Poisson(λ={lmbda})')
    plt.axvline(k, color='red', linestyle='--', label=f'k = {k}')
    plt.axhline(cdf_value, color='green', linestyle='--', label=f'P(X ≤ {k}) = {cdf_value:.2f}')
    plt.title('Cumulative Distribution Function of Poisson Distribution')
    plt.xlabel('Number of occurrences (k)')
    plt.ylabel('CDF - P(X ≤ k)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
poisson_cdf_example()
```

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
# interpret the results

import numpy as np
import scipy.stats as stats

# Define known population parameters
pop_mean = 50000  # Hypothesized population mean
pop_std = 4000    # Known population standard deviation

# Sample data (e.g., salaries of employees in a company)
sample_data = np.array([48000, 52000, 51000, 49500, 50500, 49000, 53000, 47000, 50000, 52500])

# Compute sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Calculate Z-score
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

# Compute p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference detected.")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample_mean, population_mean, std_dev, n, alpha=0.05):
    # Step 1: Calculate Z-statistic
    standard_error = std_dev / np.sqrt(n)
    z_stat = (sample_mean - population_mean) / standard_error

    # Step 2: Calculate p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Step 3: Determine critical z-value
    z_critical = norm.ppf(1 - alpha / 2)

    # Step 4: Print test results
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Critical z-value (±): ±{z_critical:.4f}")

    if abs(z_stat) > z_critical:
        print("Result: Reject the null hypothesis (significant difference).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference).")

    # Step 5: Visualization
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Decision regions
    plt.fill_between(x, 0, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region')
    plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.3)

    # Test statistic line
    plt.axvline(z_stat, color='blue', linestyle='--', linewidth=2, label=f'Z = {z_stat:.2f}')

    # Annotations
    plt.title('Two-Tailed Z-Test with Decision Regions')
    plt.xlabel('Z value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
two_tailed_z_test(sample_mean=105, population_mean=100, std_dev=15, n=50)
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(mu_0=50, mu_actual=55, sigma=10, n=30, alpha=0.05):
    """
    Visualizes Type 1 and Type 2 errors in hypothesis testing.
    - mu_0: Hypothesized population mean (null hypothesis)
    - mu_actual: Actual population mean (alternative hypothesis)
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level
    """

    # Compute critical value for rejection (Z-test)
    z_critical = stats.norm.ppf(1 - alpha)  # Right-tailed test (one-sided)
    critical_value = mu_0 + z_critical * (sigma / np.sqrt(n))

    # Generate distributions
    x = np.linspace(mu_0 - 4*sigma, mu_actual + 4*sigma, 1000)
    null_dist = stats.norm.pdf(x, mu_0, sigma / np.sqrt(n))
    alt_dist = stats.norm.pdf(x, mu_actual, sigma / np.sqrt(n))

    # Plot distributions
    plt.plot(x, null_dist, label=f"Null Distribution (μ₀={mu_0})", color="blue")
    plt.plot(x, alt_dist, label=f"Alternative Distribution (μₐ={mu_actual})", color="red")

    # Shade Type 1 Error (False Positive)
    plt.fill_between(x[x > critical_value], null_dist[x > critical_value], alpha=0.4, color="blue", label="Type 1 Error (α)")

    # Shade Type 2 Error (False Negative)
    plt.fill_between(x[x < critical_value], alt_dist[x < critical_value], alpha=0.4, color="red", label="Type 2 Error (β)")

    # Labels and legend
    plt.axvline(critical_value, color="black", linestyle="dashed", label="Critical Value")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.legend()
    plt.grid()
    plt.show()

# Call the function with default parameters
visualize_errors()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance


import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - data: list or numpy array of sample data
    - confidence: confidence level (default 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # sample standard deviation
    standard_error = sample_std / np.sqrt(n)

    # Get the t-critical value for the given confidence and degrees of freedom
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * standard_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
sample_data = [12, 15, 14, 16, 15, 14, 13, 15, 16, 14]
ci_lower, ci_upper = confidence_interval(sample_data)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
import scipy.stats as stats

def calculate_margin_of_error(sample_data, confidence_level=0.95):
    """
    Calculates the margin of error for a given confidence level using sample data.

    Parameters:
    - sample_data (array-like): The dataset (e.g., test scores, heights).
    - confidence_level (float): Confidence level (default: 95%).

    Returns:
    - margin_of_error (float): The computed margin of error.
    """

    # Compute sample statistics
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
    n = len(sample_data)

    # Compute critical value (Z-score or T-score based on sample size)
    if n > 30:  # Use Z-score if sample size > 30
        critical_value = stats.norm.ppf((1 + confidence_level) / 2)
    else:  # Use T-score if sample size is small
        critical_value = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

    # Compute margin of error
    margin_of_error = critical_value * (sample_std / np.sqrt(n))

    return margin_of_error

# Example dataset (e.g., student test scores)
sample_data = np.array([78, 85, 90, 92, 88, 76, 81, 95, 89, 84])

# Calculate margin of error
moe = calculate_margin_of_error(sample_data, confidence_level=0.95)

# Print result
print(f"Margin of Error (95% Confidence Level): {moe:.2f}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior_disease, sensitivity, specificity, positive_result=True):
    """
    Calculate the posterior probability P(Disease | Test result)

    Parameters:
    - prior_disease: P(Disease) (prevalence)
    - sensitivity: P(Positive | Disease)
    - specificity: P(Negative | No Disease)
    - positive_result: True if test result is positive, False if negative

    Returns:
    - posterior probability P(Disease | Test result)
    """
    prior_no_disease = 1 - prior_disease

    if positive_result:
        # P(Positive) = P(Positive|Disease)*P(Disease) + P(Positive|No Disease)*P(No Disease)
        p_positive = sensitivity * prior_disease + (1 - specificity) * prior_no_disease

        # Bayes' theorem:
        posterior = (sensitivity * prior_disease) / p_positive
    else:
        # P(Negative) = P(Negative|Disease)*P(Disease) + P(Negative|No Disease)*P(No Disease)
        p_negative = (1 - sensitivity) * prior_disease + specificity * prior_no_disease

        posterior = ((1 - sensitivity) * prior_disease) / p_negative

    return posterior

# Example values:
prior = 0.01          # 1% disease prevalence
sensitivity = 0.95    # 95% true positive rate
specificity = 0.90    # 90% true negative rate

posterior_prob = bayes_theorem(prior, sensitivity, specificity, positive_result=True)
print(f"P(Disease | Positive test) = {posterior_prob:.4f}")
```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
#data


import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test.

    Parameters:
    - observed: 2D numpy array or list of lists with observed frequencies

    Returns:
    - expected: 2D numpy array of expected frequencies
    """
    observed = np.array(observed)
    row_totals = observed.sum(axis=1).reshape(-1, 1)  # column vector
    col_totals = observed.sum(axis=0).reshape(1, -1)  # row vector
    grand_total = observed.sum()

    expected = (row_totals @ col_totals) / grand_total
    return expected

# Example usage:
observed_data = [
    [30, 10, 20],
    [20, 20, 20],
    [10, 30, 10]
]

expected_freq = calculate_expected_frequencies(observed_data)
print("Observed Frequencies:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(np.round(expected_freq, 2))
```
